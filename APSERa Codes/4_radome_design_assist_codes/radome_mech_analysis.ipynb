```python
import math
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science','ieee','no-latex'])
```

```python
R = 23.08/200
h = 8/100
p_atm = 1.013* (10**5) 

F = 2*(math.pi)*R*h*p_atm
'Total Force on dome: ' + str(round(F)) + ' N'
```

```python
# thickness calculator
p_cl = 2* ( 10**5) #N/m2
v = 0.35
E = 3102640781.9 #pascal

t = ( ((R**2) * p_cl * math.sqrt(3*(1-(v**2))) ) / (2*E) )**0.5
t
```

```python
p_cl_list = list(np.linspace(1.0,5.0, 20) * 10**5)
t_list = []

for i in p_cl_list:
    t_list.append((( ((R**2) * i * math.sqrt(3*(1-(v**2))) ) / (2*E) )**0.5) * 1000)

plt.plot(t_list, [x / 1000000 for x in p_cl_list], color='m')
plt.axvline(0.8345111725512006, color='y', linestyle='--', linewidth = 0.5)
plt.axhline(0.2, color='y', linestyle='--', linewidth = 0.5, label = 't = 0.8 mm')

plt.title("Plexiglass")
plt.xlabel('Thickness (mm)')
plt.ylabel('Buckling Pressure (Mpa)')
plt.legend(loc="best")
plt.savefig('bucklingpressure_thickness_radome_plexiglass.pdf', dpi=500)
```

```python
# thickness calculator
p_cl = 2* ( 10**5) #N/m2
v = 0.35
E = 3102640781.9 #pascal
R = 23.08/200
t = 0.8
w_max = (3*p_cl*(R**4))/(16*E*(1-(v**2))*(t**3))
w_max*1000000
```

```python
t_list_w_max_plexiglass = list(np.linspace(0.55, 1.3, 20 ))
w_max_list = []

for i in t_list_w_max_plexiglass:
    w_max_list.append((3*p_cl*(R**4))/(16*E*(1-(v**2))*(i**3)))
#[x*1000000 for x in t_list_w_max_plexiglass]
plt.plot(t_list_w_max_plexiglass,[x*1000000 for x in w_max_list], color='m')
#plt.axvline(0.8345111725512006, color='y', linestyle='--', linewidth = 0.5)
#plt.axhline(0.2, color='y', linestyle='--', linewidth = 0.5, label = 't = 0.8 mm')

plt.title("Plexiglass")
plt.xlabel('Thickness (mm)')
plt.ylabel('Maximum deflections (um)')
plt.legend(loc="best")
plt.savefig('max_deflection_radome_plexiglass.pdf', dpi=500)
```
