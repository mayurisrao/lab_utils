```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from scipy import stats
```

```python
### Read values from txt file
volt1=np.array([]) 
volt2=np.array([]) 
#counter=0
with open("run_09032024_pt100.txt", "r") as filestream:
    for line in filestream:
        #print(counter)
        fields=line.split(",")
        v1=float(fields[3])
        volt1 = np.append(volt1,v1) 
        v2=float(fields[7])
        volt2 = np.append(volt2,v2)
        #counter+=1
        #print(fields[7])
        #if(np.remainder(counter,1000)==0):
        #    print(counter)
time=np.linspace(0,len(volt1),len(volt1))
#print(time)
plt.plot(time,volt1)

```

```python
plt.plot(time,volt2)
```

```python
resis1=150*(volt1)/(5-(volt1))
resis2=150*(volt2)/(5-(volt2))
plt.plot(time,resis1,label='hot run')
plt.plot(time,resis2,label='cold run')
plt.xlabel("Time (arbitrary units)")
plt.ylabel("Resistance (ohms)")
plt.title("08/03/2024 PT100 arduino data")
plt.grid()
plt.legend()
plt.show()
```

```python
#R=R0(1+At+BT**2)
r0=125
A=3.98*10**(-3)
B=-5.87*10**(-7)
coeff1=[[0 for x in range(3)] for x in range(len(resis1))]
sol1=[[0 for x in range(2)] for x in range(len(resis1))]
for i in range(len(resis1)):
    coeff1[i]=np.append(coeff1[i],[1,A/B,(1-resis1[i]/r0)*(1/B)])
    sol1[i]=(np.roots(coeff1[i]))
```

```python
temp1=np.zeros(len(sol1))
for i in range(len(sol1)):
    temp1[i]=sol1[i][1]
#print(temp1[3])
plt.plot(time,temp1)
```

```python
#R=R0(1+At+BT**2) where c=0 for temp>0 deg C
# RTD/ PT100 eq: R = R0 [1 + A t + B t^2 + C (t  100) t^3]
r0=125
A=3.98*10**(-3)
B=-5.87*10**(-7)
coeff2=[[0 for x in range(3)] for x in range(len(resis2))]
sol2=[[0 for x in range(2)] for x in range(len(resis2))]
for i in range(len(resis2)):
    coeff2[i]=np.append(coeff2[i],[1,A/B,(1-resis2[i]/r0)*(1/B)])
    sol2[i]=(np.roots(coeff2[i]))
```

```python
temp2=np.zeros(len(sol2))
for i in range(len(sol2)):
    temp2[i]=sol2[i][1]
#print(temp1[3])
plt.plot(time,temp2)
```

```python
plt.plot(time,temp1,label='hot run')
plt.plot(time,temp2,label='cold run')
plt.xlabel("Time (arbitrary units)")
plt.ylabel("Temperature in deg")
plt.title("08/03/2024 PT100 arduino data")
plt.grid()
plt.legend()
plt.show()
```
